# Exploratory Data Analysis

## Most Common Words

### Analysis

```python
import pandas as pd

data = pd.read_pickle('dtm.pkl')
data = data.transpose()
data.head()
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c]= list(zip(top.index, top.values))

top_dict
```

```python
for poet, top_words in top_dict.items():
    print(poet)
    print(', '.join([word for word, count in top_words[0:14]]))
    print('---')
```

```python
from collections import Counter

words = []
for poet in data.columns:
    top = [word for (word, count) in top_dict[poet]]
    for t in top:
        words.append(t)
        
words
```

```python
Counter(words).most_common()
```

```python
add_stop_words = [word for word, count in Counter(words).most_common() if count > 6]
add_stop_words
```

```python
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer

data_clean = pd.read_pickle('data_clean.pkl')

stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index

import pickle
pickle.dump(cv, open("cv_stop.pkl", "wb"))
data_stop.to_pickle("dtm_stop.pkl")
```

```python
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16, 6]

full_names = ['Allen Ginsberg', 'Axentioi Alexandru', 'Christina Georgina Rossetti', 'George Gordon Byron', 'Iff Ur Abs','Judith Wright',
              'Max Ehrmann', 'Nuno Cardoso', 'Paul Laurence Dunbar', 'Sammeh Love']

for index, poet in enumerate(data.columns):
    wc.generate(data_clean.transcript[poet])
    
    plt.subplot(3, 4, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(full_names[index])
    
plt.show()
```

### Findings

* Sammeh Love uses the f-word a lot
* Nuno Cardoso uses the s-word a lot 

## Number of words

## Analysis

```python
unique_list = []
for poet in data.columns:
    uniques = data[poet].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

data_words = pd.DataFrame(list(zip(full_names, unique_list)), columns=['poet', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

```python
total_list = []
for poet in data.columns:
    totals = sum(data[poet])
    total_list.append(totals)

run_times = [60, 59, 80, 60, 67, 73, 77, 63, 62, 58]

data_words['total_words'] = total_list
data_words['run_times'] = run_times
data_words['words_per_minute'] = data_words['total_words'] / data_words['run_times']

data_wpm_sort = data_words.sort_values(by='words_per_minute')
data_wpm_sort
```

```python
import numpy as np

y_pos = np.arange(len(data_words))

plt.subplot(1, 2, 1)
plt.barh(y_pos, data_unique_sort.unique_words, align='center')
plt.yticks(y_pos, data_unique_sort.poet)
plt.title('Number of Unique Words', fontsize=20)

plt.subplot(1, 2, 2)
plt.barh(y_pos, data_wpm_sort.words_per_minute, align='center')
plt.yticks(y_pos, data_wpm_sort.poet)
plt.title('Number of Words Per Minute', fontsize=20)

plt.tight_layout()
plt.show()
```

### Findings

* **Vocabulary**
  * Allen Ginsberg uses a lot of unique words in his poems
  * Christina Georgina Rossetti has the smallest vocabulary
* **Talking speed**
  * Allen Ginsberg talks fast
  * Christina Georgina Rossetti talks slow
  
Nuno Cardoso is somewhere in the middle in both cases

```python
data_bad_words = data.transpose()[['fucking', 'fuck', 'shit']]
data_profanity = pd.concat([data_bad_words.fucking + data_bad_words.fuck, data_bad_words.shit], axis=1)
data_profanity.columns = ['f_word', 's_word']
data_profanity
```

```python
plt.rcParams['figure.figsize'] = [10, 8]

for i, comedian in enumerate(data_profanity.index):
    x = data_profanity.f_word.loc[comedian]
    y = data_profanity.s_word.loc[comedian]
    plt.scatter(x, y, color='blue')
    plt.text(x+1.5, y+0.5, full_names[i], fontsize=10)
    plt.xlim(-5, 155) 
    
plt.title('Number of Bad Words Used in Routine', fontsize=20)
plt.xlabel('Number of F Bombs', fontsize=15)
plt.ylabel('Number of S Words', fontsize=15)

plt.show()
```

### Findings

* **Use of F-words** - Sammeh Love has used the most F-words in their poem, which is 12!
* **Clean poem** - Poets besides Sammeh, Axentioi and Nuno write clean poems.

Here, I have compared the use of old words used by the poets in their poems. Example - "thou" and "thee"

```python
data_old_words = data.transpose()[['thou', 'thee']]
data_old_words
```

### Findings

* George Gordon Byron has used 'thou' 1 time and 'thee' 6 times
* Paul Laurence Dunbar has used 'thee' 1 time in their poem
* Poets other than George Gordon Byron and Paul Laurence Dunbar have not used old words in their poem.

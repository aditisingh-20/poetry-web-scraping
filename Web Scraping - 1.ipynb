## Getting the Data

```python
import requests
from bs4 import BeautifulSoup
import pickle

def url_to_poem(url):
    '''Returns transcript data specifically from scrapsfromtheloft.com.'''
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    '''<div class="preview poem_body"><div class="tr_16976251"></div><div class="orig_16976251">loose leaves fall<br>'''
    #text = [p.text for p in soup.find(class_="preview poem_body").find_next(text=True).next]
    tag=soup.find("div", {"class": "preview poem_body"})
    text=''
    for string in tag.strings:
        text=text+string
    print(url)
    return text

urls = ['https://allpoetry.com/poem/16976251-life-and-death-by-Iff-Ur-Abs',
        'https://allpoetry.com/-A-Prayer-',
        'https://allpoetry.com/poem/16773295-Sometimes-it-could-be-faster--by-Nuno-Cardoso',
        'https://allpoetry.com/poem/12183098-ing-I.-by-Sammeh-Love-adult',
        'https://allpoetry.com/When-We-Two-Parted',
        'https://allpoetry.com/We-Wear-the-Mask',
        'https://allpoetry.com/Hurt-No-Living-Thing',
        'https://allpoetry.com/poem/8521455-Woman-To-Man-by-Judith-Wright',
        'https://allpoetry.com/poem/8495157-Howl-by-Allen-Ginsberg-adult',
        'https://allpoetry.com/poem/16893629-What-is-poetry--by-Axentioi-Alexandru'
        ]

poets = ['iff','max','nuno','sammeh','george','paul','christina','judith','allen','axentioi']
```

```python
transcripts = [url_to_poem(u) for u in urls]
```

```python
!mkdir transcripts

for i, c in enumerate(poets):
    with open("transcripts/" + c + ".txt", "wb") as file:
        pickle.dump(transcripts[i], file)
```

```python
next(iter(data.keys()))
```

```python
def combine_text(list_of_text):
    '''Takes a list of text and combines them into one large chunk of text.'''
    combined_text = ''.join(list_of_text)
    return combined_text
```

```python
data_combined = {key: [combine_text(value)] for (key, value) in data.items()}
```

```python
import pandas as pd
pd.set_option('max_colwidth',150)

data_df = pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns = ['transcript']
data_df = data_df.sort_index()
data_df
```

```python
data_df.transcript.loc['iff']
```

```python
import re
import string

def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data_df.transcript.apply(round1))
data_clean
```

```python
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.transcript.apply(round2))
data_clean
```

1. Can you add an additional regular expression to the clean_text_round2 function to further clean the text?

Here, I have added the regular expression '–' to further clean the text.

```python
def clean_text_round2(text):
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('–', '', text) # Added regular expression '–'
    return text

round2 = lambda x: clean_text_round2(x)

data_clean = pd.DataFrame(data_clean.transcript.apply(round2))
data_clean
```

2. Play around with CountVectorizer's parameters. What is ngram_range? What is min_df and max_df?

ngram_range is the lower and upper boundary of the range of n-values for different word n-grams or char n-grams to be extracted. Increasing the ngram_range will mean the vocabulary is expanded from single words to short phrases of your desired lengths.

min_df ignores the terms that have a document frequency strictly lower than the given threshold. max_df ignores the terms that have a document frequency strictly higher than the given threshold.

Here, I have increased the ngram_range value to (2,2) for the CountVectorizer which means it will return bigrams (2-grams) or two word phrases. 

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english',ngram_range=(2,2))
data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```

Here, I have set the value of min_df to 2 and max_df to 100 for CountVectorizer, which means terms that only have document frequency between 2 and 100 will be displayed

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english',min_df=2, max_df=100)
data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```

Here, I have set the value of Binary to True for the CountVectorizer. This means that all non zero counts are set to 1

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english',binary=True)
data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```
